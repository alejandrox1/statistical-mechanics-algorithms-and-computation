Given an ordered set, a permutation is a (bijection) reordering of its elements.
Combinations on the other hand are subsets of the given set without regard to the order of its elements.

The number of permutations of n distinct objects is $n!$.

Using Cauchys two line notation, one lists the elements of set $S$ in the first row, and for each one, its  image below in the second row. 
For example, a permutation of the set $S = \{ 1,2,3 \}$ can be depicted as:
$$
\sigma = \begin{bmatrix}
    1 & 2 & 3  \\
    3 & 2 & 1 \\
\end{bmatrix}
$$

so $\sigma(1) = 3$. The elements of the set $S$ (the first row) may appear in any order. So for example, the previous permutation can also be written as:
$$
\sigma = \begin{bmatrix}
    3 & 1 & 2  \\
    1 & 3 & 2 \\
\end{bmatrix}
$$
and thus, $\sigma(1) = 3$, yet again.

```python
N=100
pi_max = max(prob.values())

counter = prob.copy()
counter = {x:0 for x in counter}
activities = [x for x in prob.keys() ]

for throw in range(N):
    k = random.randint(0,len(activities)-1)
    p = random.uniform(0,pi_max)
    if (p<=prob[activities[k]]):
        counter[activities[k]] += 1

X = np.arange(len(prob))
plt.bar(X, counter.values(), align='center', width=0.5)
plt.xticks(X, prob.keys())
plt.show();
print(counter)
```

$$ \{k, \pi_k\} \rightarrow \{x, \pi(x)\} $$

$$
\Pi _k = \Pi_{k-1} + \pi_k   \rightarrow   \Pi(x) = \int_{-\infty}^{x} dx^\prime \, \pi (x^\prime)
$$
where $ \Pi(x) = \Pi(x-dx) + \pi(x)dx$

Discrete distributions: find $k$ with $\Pi_{k-1} < \Gamma < \Pi_k$

Continuous distribution: find $x$ with $\Pi(x) = \Gamma$.
Here $x = \Pi^{-1}(\Gamma)$
